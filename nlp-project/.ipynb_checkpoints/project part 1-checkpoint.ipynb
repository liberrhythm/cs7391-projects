```python
delchars += '—'
```

```python
adjective = ['JJ', 'JJR', 'JJS']
verb = ['VB', 'VBD', 'VBG', 'VBN', 'VBP', 'VBZ']
noun = ['NN', 'NNP', 'NNPS', 'NNS',]
adverb = ['RB', 'RBR', 'RBS']
# allowed_tag_list = adjective + verb + noun + adverb
allowed_tag_list = verb + noun
```

```python
speech_sent_dict = {}
```

```python
for speech in df.speech_id.unique():
    sentences = list(df[df["speech_id"] == speech]["text"])
    sentences = [word_tokenize(sent.lower()) for sent in sentences]
    new_sentences = []
    for sent in sentences:
        sent = [token.translate(str.maketrans('', '', delchars)) for token in sent if len(token) > 1]
        pos = pos_tag(sent)
        dic = dict(pos)
        new_sent = [token for token in sent if token not in swlist and dic[token] in allowed_tag_list]
        new_sentences.append(new_sent)
    tokens = [token for sent in new_sentences for token in sent]
    print(tokens)
    speech_sent_dict[speech] = tokens
```
